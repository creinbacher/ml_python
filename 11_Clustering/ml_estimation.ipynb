```python
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def generate_dataset():
    var = 3
    mean = 1.5
    data = np.random.normal(mean, var, 10)
    return data
```

```python
data = generate_dataset()

plt.scatter(data, [0 for i in range(len(data))])
plt.show()
```

```python
def estimate_mean(data):
    mean = np.mean(data)
    return mean
```

```python
def estimate_std(data):
    std = np.std(data)
    return std
```

```python
mean = estimate_mean(data)
std = estimate_std(data)

print(f"Mean: {mean}")
print(f"Std: {std}")
```

```python
real_pdf = [norm.pdf(x, 1.5, 3) for x in np.arange(-0, 11, 0.01)]
estimate_pdf = [norm.pdf(x, mean, std) for x in np.arange(-0, 11, 0.01)]
```

```python
plt.scatter(np.arange(-0, 11, 0.01), real_pdf, color="red", s=4)
plt.scatter(np.arange(-0, 11, 0.01), estimate_pdf, color="blue", s=4)
plt.legend(["Y", "$\hat{Y}$"])
plt.show()
```

```python
for num_samples in range (10, 100, 10):
    data = np.random.normal(1.5, 3, num_samples)
    mean = estimate_mean(data)
    std = estimate_std(data)
    real_pdf = [norm.pdf(x, 1.5, 3) for x in np.arange(-0, 11, 0.01)]
    estimate_pdf = [norm.pdf(x, mean, std) for x in np.arange(-0, 11, 0.01)]
    plt.scatter(np.arange(-0, 11, 0.01), real_pdf, color="red", s=4)
    plt.scatter(np.arange(-0, 11, 0.01), estimate_pdf, color="blue", s=4)
    plt.legend(["Y", "$\hat{Y}$"])
    plt.title("Num Samples = "+str(num_samples))
    plt.show()
```
